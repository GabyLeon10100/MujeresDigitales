## Ejercicio 1:

Cargue los datos del archivo ```pokemon_data.csv``` a un dataframe, cree una columna que se llame interes asignándole el valor False, luego cambie el valor de esta columna solo para los pokemon que sean legendarios (Legendary = True) y que el Type 1 sea fuego.

````python
import pandas as pd
````

````python
#Traemos el archivo "pokemon_data.csv"
from google.colab import files
uploaded = files.upload()
````

````python
#Mostramos el archivo
import io 
df_pokemon = pd.read_csv(io.BytesIO(uploaded['pokemon_data.csv']), sep=",")
df_pokemon.head(10)
````

````python
#Creamos el campo "interes", con valor predeterminado como False
df_pokemon['interes'] = False
df_pokemon.head(10)
````

````python
#Recorremos el valor que deseamos para cambiar el valor de interes
#Los campos que se quieren son: Legendary = True y Type = Fire
mask1 = df_pokemon['Type 1'] == "Fire" 
mask2 = df_pokemon['Legendary'] == True
df_final = df_pokemon[mask1&mask2] #Creamos un dataFrame secundaria para obtener los campos que cumplen con ambas condiciones
mask_final = df_final.index #Traemos los indices, para poderlo modificar en el dataframe principal
````

````python
df_pokemon.loc[mask_final, 'interes'] = True #Realizamos el cambio del valor de interes a true a los campos que deseamos
df_pokemon.head(800)
````

````python
#Filtramos para validar que si haya realizado los cambios
df_pokemon[df_pokemon['interes'] == True]
````
